```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid=pd.read_csv('ECDC_COVID_19.csv')
covid
```

```python
top_death_rate=covid.nlargest(5,'deaths')
top_death_rate
```

```python
top_5=top_death_rate.loc[:,['deaths','countriesAndTerritories']]
top_5
```

```python
top_death_rate=covid.groupby('countriesAndTerritories')[['deaths','cases']].sum()
top_death_rate.reset_index
```

```python
top_5=top_death_rate.sort_values(by='deaths',ascending=False)
top_5
```

```python
top=top_5.reset_index()
top=top.head(5)
top
```

#### Top countries with most deaths

```python
plt.title('Top countries with most deaths')
plt.plot(top['countriesAndTerritories'],top['deaths'],color='r')
plt.xlabel('Countries')
plt.ylabel('Death rate')
plt.show()
```

#### top countries with cases and death

```python
plt.title('Top countries with cases and death')
plt.bar(top['countriesAndTerritories'],top['cases'])
plt.xlabel('Country')
plt.ylabel('Death rate & Case rate')
plt.xticks(rotation=90)
plt.show()
```
